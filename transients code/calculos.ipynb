```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

path = '../resultados/prueba.csv'
csv = pd.read_csv(path, delimiter=',')

csv
```

```python
Tipo = csv['Tipo'].unique()
```

### Violin plot using median

```python
import seaborn as sns

sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(11, 6))

# ax.set_title('Violin Plot of Alpha by Tipo', fontsize=16, pad=20)

palette = sns.color_palette("viridis", as_cmap=True)

sns.violinplot(data=csv, x="Tipo", y="Alpha", orient="x", bw_adjust=.5, cut=1, linewidth=1, palette="Set3")
sns.despine(left=True, bottom=True)
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(11, 6))
sns.violinplot(csv, x="Alpha", y="Tipo", orient="y", bw_adjust=.5, cut=1, linewidth=1, palette= "Set3")
sns.despine(left=True, bottom=True)
```

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(11, 6))

palette = sns.color_palette("viridis", as_cmap=True)

sns.swarmplot(data=csv, x="Tipo", y="Alpha", palette = 'tab10', size=4,)
sns.despine(left=True, bottom=True)
```

### violin plot using mean

```python
import seaborn as sns
from numpy import mean



sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(11, 6))

# ax.set_title('Violin Plot of Alpha by Tipo', fontsize=16, pad=20)

palette = sns.color_palette("viridis", as_cmap=True)

sns.violinplot(data=csv, x="Tipo", y="Alpha", orient="x", bw_adjust=.5, cut=1, linewidth=1, palette="Set3")
sns.pointplot(x = 'Tipo', y='Alpha', data=csv, estimator=mean)

sns.despine(left=True, bottom=True)
```
